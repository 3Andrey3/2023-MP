```python
#список из 99999 случайных вещественных чисел 
random_list = [random.uniform(-1,1) for i in range(10000)]
print("Список случайных вещественных чисел:", random_list)

#42000 разных точки комплексной плоскости
birth_day = 1 
birth_month = 1 
r = birth_day/birth_month
n=10000
def PointsInCircum(r,n):
    return [(math.cos(2*math.pi/n*x)*r,math.sin(2*math.pi/n*x)*r) for x in range(0,n+1)]
t = PointsInCircum(r,n)
random.shuffle (t)
print("Точки на комлексной плоскости:", t)
```

```python
file = 'book.txt'
book = []
with open(file, encoding='utf-8') as file:
    for line in file:
        for i in line.split():
            book.append(i.strip(string.punctuation))
print("Отрывок из книги:", book[:10000])
```

```python
def bubble_sort(nums):  
    # Устанавливаем swapped в True, чтобы цикл запустился хотя бы один раз
    swapped = True
    while swapped:
        swapped = False
        for i in range(len(nums) - 1):
            if nums[i] > nums[i + 1]:
                nums[i], nums[i + 1] = nums[i + 1], nums[i]
                swapped = True

bubble_sort(list_integers)  
print("№2")
print(list_integers)
```

```python
def selection_sort(arr):
    for i in range(len(arr)):
        minz, ind = i, i
        for j in range(i, len(arr)):
            if arr[j] < arr[minz]:
                minz = j
                ind = j
        arr[i], arr[ind] = arr[ind], arr[i]
    return arr
kl=selection_sort(book[:10000])
print("№13")
print(kl)
```

```python
def insertion_sort(fr):
    for i in range(1, len(fr)):
        paste_element = fr[i]
        while i > 0 and fr[i-1] > paste_element:
            fr[i] = fr[i-1]
            i = i-1
        fr[i] = paste_element
    return fr
cv= insertion_sort(t)
print("№4")
print(cv)
```
